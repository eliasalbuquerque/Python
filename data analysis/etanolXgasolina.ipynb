## Tratamento de dados <a class = "anchor" id = "tratamento" ></a>

```python
# remoção das colunas desnecessárias;
print('[1]')
tabela = pd.read_csv(r'C:\Users\elias\Downloads\20210508_135821_DRIVVO.csv', encoding = 'latin1', skiprows = 1)
print(tabela.info(),'\n')

print('[2]')
tabela = tabela.drop(['Data', 'Completou o tanque', 'Segundo combustível', 'Preço / L.1', 'Valor total.1', 
                      'Volume.1', 'Completou o tanque 2', 'Terceiro combustível', 'Preço / L.2', 'Valor total.2', 'Volume.2', 
                      'Completou o tanque 3', 'Posto de combustível', 'Motivo', 'Observação'], axis = 1)
print(tabela.info())
display(tabela)


# selecionar apenas linhas com os combustíveis;
print('[3]')
# tabela[tabela['Combustível'] == 'Etanol']    #-------> Só funciona para um argumento;
tabela = tabela[tabela['Combustível'].isin(['Etanol', 'Gas. Aditivada', 'Gas. Comum'])]
display(tabela)

'''
para uma lista de argumentos:
    .isin(["lista de argumentos"])
    Lê-se: na tabela, na coluna Combustível da tabela, filtrar "Etanol", "Gas. Aditivada" e "Gas. Comum"
    
ou "string contains":
    df[df['col3'].str.contains('py|dat|\+')]    
    Lê-se: no dataframe, na col3 do dataframe, filtrar strings que contém: "py", "dat", "+"
'''


# transformando os "Dtype" das colunas em números inteiros e decimais;
print('[4]')
# tabela['Odômetro (km)'] = pd.to_numeric(tabela['Odômetro (km)'], errors = 'coerce') # apenas para uma coluna;
tabela = tabela.astype({'Odômetro (km)':'int32', 'Preço / L':'float16', 'Valor total':'float16', 'Volume':'float16'})

tabela.dtypes
display(tabela)
```

### Considerações

- Foi constatado que as colunas com os dados a serem considerados na análise são submenus da coluna "#Abastecimento" e por isso o código `display(tabela.columns)` na extração dos dados apresentou um comportamento extranho:
- Corrigido com o parâmetro `skiprowns = 1` (pula a primeira linha na extração dos dados) do `pd.read_csv`. O padrão é `skiprows=None`;
- Para remover várias colunas, tem que colocar `.drop(['colunas1', 'coluna2'], axis = 1)`;
- Foi feita a seleção somente dos combustíveis e a transformação dos tipos de dados de cada coluna;
